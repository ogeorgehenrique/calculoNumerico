```python


# Metodo que define a Função f(x)
def f(x):
    return x**2 - 2

# Metodo que define a Derivada da Função f(x)
def df(x):
    return 2*x


# Metodo de Newton
def newton_raphson(x0, tolerancia, max_iteracoes):
    print(f"=========================================================")
    print(f"Iniciando Newton-Raphson com chute inicial: {x0}")
    print(f"=========================================================")

    x_atual = x0

#Inicio do laço for com i = 0 continue no laço ate que o limite de interaçoes max_iteracoes
    for i in range(max_iteracoes):

        # 1. Calcular valores da função e da derivada
        fx = f(x_atual)
        dfx = df(x_atual)


        # PRECAUÇÃO: Derivada zero = divisão por zero (linha reta horizontal)
        if dfx == 0:
            print("Erro: Derivada zero. O método falhou.")
            return None

        # 2. Aplicar a fórmula de Newton
        x_novo = x_atual - (fx / dfx)

        print("=========================================================")
        print(f"Iteração {i+1}")
        print(f"F(x_atual) = {fx:.8f}")
        print(f"dF(x_atual) = {dfx:.8f}")
        print(f"x_atual = {x_atual:.8f} -> x_novo = {x_novo:.8f}")

        # 3. Verificar critério de parada (Erro relativo ou absoluto)
        if abs(x_novo - x_atual) < tolerancia:
            print(f"x_novo - x_atual = {x_novo - x_atual:.8f} < tolerancia = {tolerancia}")
            return x_novo


        print(f"x_novo - x_atual = {x_novo - x_atual:.8f} > tolerancia = {tolerancia}")
        # Atualiza para o próximo loop
        x_atual = x_novo

    print("Erro: Número máximo de iterações atingido.")
    return None

# --- Executando ---
# Chute inicial: 3 (longe da raiz que é 1.41...)
print("Função: F(x)= x^2 -2")
print("Derivada da Função: F(x)= 2x")
print("")



#chamada do metordo de Newton
raiz = newton_raphson(3, 0.0001, 100)

print("=========================================================")
print(f"\nA raiz encontrada é: {raiz:.8f}")
print(f"Valor da função na raiz: f(raiz) = {f(raiz):.16f}")
```
